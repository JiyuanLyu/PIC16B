```python
import networkx as nx
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
results = pd.read_csv("pagerank/pages.csv")
G = nx.from_pandas_edgelist(results, source = "source", target = "target", create_using = nx.DiGraph)
results = results[results["move_type"] == "walk"]
```

```python
ranks = nx.degree(G)
fig, ax = plt.subplots(1, figsize = (15, 10))
pos = nx.fruchterman_reingold_layout(G)
nx.draw(G, 
        pos, 
        node_size= 20*np.array([ranks[v] for v in G.nodes()]), 
        with_labels = True, 
        edge_color="lightgrey",
        font_size = 6,
        node_color = "#F5B895",
        arrows = True)

plt.savefig("pagerank_viz.png", dpi = 300)
```
